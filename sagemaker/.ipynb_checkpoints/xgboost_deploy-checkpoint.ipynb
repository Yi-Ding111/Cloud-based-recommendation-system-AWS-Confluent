```python
#this is the learning model process using data from ETL, choosing XGBoost
```

```python
import pandas as pd
import os
import boto3
import sagemaker
from sagemaker import get_execution_role
from sagemaker.xgboost.estimator import XGBoost
from sagemaker.tuner import IntegerParameter, ContinuousParameter, HyperparameterTuner
```

```python
#read data from s3
```

```python
data_position='s3://*****/output_small/data_small.csv'
data=pd.read_csv(data_position)
```

```python
train_data_position='s3://*****/data/order_products/order_products__train.csv.gz'
train_data=pd.read_csv(train_data_position)
```

```python
#order info data
```

```python
info_order_position='s3://*****/data/orders/orders.csv'
info_order=pd.read_csv(info_order_position)
```

```python
#preview dataset
```

```python
train_data.head(10)
```

```python
data.head()
```

```python
info_order.head()
```

```python
#data pre-processing
```

```python
#left join user_id in infto_order to order_id in train_data
```

```python
train_data_v1=pd.merge(train_data,info_order[['order_id','user_id']],how='left',on='order_id')
```

```python
train_data_v1.head(10)
```

```python
train_data_v1.shape
```

```python
pd.isnull(train_data_v1['user_id']).values.any()
```

```python
# left join order_id in info_order to data on order_id
```

```python
data_v1=pd.merge(data,info_order[info_order.eval_set != 'prior'][['user_id','eval_set']],how='left',on='user_id')
```

```python
data_v1.head()
```

```python
pd.isnull(data_v1['eval_set']).values.any()
```

```python
#data_v2=pd.merge(data_v1,train_data_v1[['user_id','product_id','reordered']],how='left',on=['user_id','product_id'])
data_v2 = data_v1.merge(train_data_v1[['user_id', 'product_id', 'reordered']], how = 'left')
```

```python
data_v2.head()
```

```python
#check it existing NULL value 
```

```python
print(pd.isnull(data_v2[['reordered']]).values.any())
data_v2.shape
```

```python
data_v2[data_v2[['reordered']].isnull().T.any()]
```

```python
#because too many records have reordered value with NAN, means that consumer may did not order again ,so fill 0
data_v2['reordered']=data_v2['reordered'].fillna(0)
data_v2['reordered']=data_v2.reordered.astype(int)
```

```python
pd.isnull(data_v2[['reordered']]).values.any()
```

```python
#feature engineering
```

```python
data_v2['prod_reorder_probability'] = data_v2.prod_second_orders / data_v2.prod_first_orders
data_v2['prod_reorder_times'] = 1 + data_v2.prod_reorders / data_v2.prod_first_orders
data_v2['prod_reorder_ratio'] = data_v2.prod_reorders / data_v2.prod_orders
#delete unnecessary col
data_v2.drop(['prod_reorders', 'prod_first_orders', 'prod_second_orders'], axis=1, inplace=True)
data_v2['user_average_basket'] = data_v2.user_total_products / data_v2.user_orders
data_v2['up_order_rate'] = data_v2.up_orders / data_v2.user_orders
data_v2['up_orders_since_last_order'] = data_v2.user_orders - data_v2.up_last_order
data_v2['up_order_rate_since_first_order'] = data_v2.up_orders / (data_v2.user_orders - data_v2.up_first_order + 1)
```

```python
data_v2.head()
```

```python
#split data for training and test
```

```python
train = data_v2[data_v2.eval_set == 'train'].copy()
test = data_v2[data_v2.eval_set == 'test'].copy()
```

```python
test_gt = data_v2[data_v2.eval_set == 'test'].copy()
test_gt=test_gt[['reordered']]
test_gt=test_gt.reset_index().drop(['index'],axis=1)
test_gt.head()
```

```python
#drop un-necessary col
```

```python
test.drop(['product_id','user_id','eval_set','reordered'],axis=1,inplace=True)
train.drop(['product_id','user_id','eval_set'],axis=1,inplace=True)
```

```python
#split variables and turth
```

```python
train_x=train.drop(['reordered'],axis=1)
train_y=train[['reordered']]
```

```python
#split training set and validation set
```

```python
train_xgb_x=train_x[20000:]
train_xgb_y=train_y[20000:]

val_xgb_x=train_x[:20000]
val_xgb_y=train_y[:20000]
```

```python
#store the processed data into data lake: s3
```

```python
if not os.path.exists('data/xgboost'):
    os.makedirs('data/xgboost')

#save dataframes into csv
pd.DataFrame(test).to_csv(os.path.join('data/xgboost','test.csv'),header=False,index=False)
pd.concat([val_xgb_y,val_xgb_x],axis=1).to_csv(os.path.join('data/xgboost','validation.csv'),header=False,index=False)
pd.concat([train_xgb_y,train_xgb_x],axis=1).to_csv(os.path.join('data/xgboost','train.csv'),header=False,index=False)
```

```python
#release memory
```

```python
val_xgb_x=val_xgb_y=train_xgb_x=train_xgb_y=None
```

```python
session=sagemaker.Session()
test_location = session.upload_data(os.path.join('data/xgboost', 'test.csv'), key_prefix='xgboost-data')
val_location = session.upload_data(os.path.join('data/xgboost', 'validation.csv'), key_prefix='xgboost-data')
train_location = session.upload_data(os.path.join('data/xgboost', 'train.csv'), key_prefix='xgboost-data')
```

```python
#classification
```

```python
#give permission to access data

role=get_execution_role()
print(role)
```

```python
#set xgboost hyperparameters
```

```python
'''
hyperparameters = {
        "max_depth":"5",
        "eta":"0.1",
        "gamma":"4",#Minimum drop in loss function required for node splitting
        "min_child_weight":"6",
        "subsample":"0.8",#random sampling
        "objective":"binary:logictic", #we want probability as output
        "num_round":"50"}
'''
```

```python
#output path where the trained model

bucket = sagemaker.Session().default_bucket()
prefix = 'xgboost-train'
output_path = 's3://{}/{}/output'.format(bucket, prefix)
```

```python
#looks for the XGBoost image URI

from sagemaker import image_uris
xgboost_container = sagemaker.image_uris.retrieve("xgboost", session.boto_region_name, "latest")
```

```python
#create xgboost estimator

xgb_model = sagemaker.estimator.Estimator(image_uri=xgboost_container, 
                                          #hyperparameters=hyperparameters,
                                          role=role, 
                                          instance_count=1, #num of compute instance
                                          instance_type='ml.m5.2xlarge',
                                          output_path=output_path,
                                          sagemaker_session=session)
```

```python
xgb_model.set_hyperparameters(max_depth=5,
                        eta=0.1,
                        gamma=4,
                        min_child_weight=6,
                        subsample=0.8,
                        silent=0,
                        objective='binary:logistic',
                        early_stopping_rounds=10,
                        num_round=500)
```

```python
# create hyperparameter tuner to optimize model
```

```python
xgb_hyperparameter_tuner = HyperparameterTuner(estimator = xgb_model,
                                               objective_metric_name = 'validation:rmse',
                                               objective_type = 'Minimize', # Whether we wish to minimize or maximize the metric.
                                               max_jobs = 4, # The total number of models to train
                                               max_parallel_jobs = 2, # The number of models to train in parallel
                                               hyperparameter_ranges = {
                                                    'max_depth': IntegerParameter(3, 12),
                                                    'eta'      : ContinuousParameter(0.05, 0.5),
                                                    'min_child_weight': IntegerParameter(2, 8),
                                                    'subsample': ContinuousParameter(0.5, 0.9),
                                                    'gamma': ContinuousParameter(0, 10),
                                               })
```

```python
# fit the xgb model with tuner

train_data = sagemaker.TrainingInput(s3_data=train_location, content_type='csv')
val_data = sagemaker.TrainingInput(s3_data=val_location, content_type='csv')
xgb_hyperparameter_tuner.fit({'train': train_data, 'validation': val_data})
```

```python
#choose the parameter group with best performance and attach model
```

```python
xgb_final=sagemaker.estimator.Estimator.attach(xgb_hyperparameter_tuner.best_training_job())
```

```python
#create transformer object to do interactions with the trained model
```

```python
xgb_transformer=xgb_final.transformer(instance_count = 1,instance_type='ml.m4.xlarge')
```

```python
#test the transformer job, do interactions with test data
```

```python
xgb_transformer.transform(test_location,content_type='text/csv',split_type='Line')
```
